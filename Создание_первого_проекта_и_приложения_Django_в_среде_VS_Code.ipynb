### 1. Введение в Django

**Django** — это мощный фреймворк для веб-разработки на Python, который позволяет быстро создавать безопасные и масштабируемые веб-приложения. Django следует принципу DRY (Don't Repeat Yourself) и предоставляет множество встроенных инструментов для работы с базами данных, аутентификацией, администрированием и т.д.



### 2. Установка необходимого ПО

Для работы с Django в VS Code вам понадобится:

1. **Python** (версия 3.7 и выше) — [скачать с официального сайта](https://www.python.org/downloads/).
2. **Visual Studio Code** — [скачать с официального сайта](https://code.visualstudio.com/).
3. **Расширение Python для VS Code** — установите через Marketplace в VS Code.



### 3. Создание виртуального окружения

Виртуальное окружение позволяет изолировать зависимости вашего проекта от других проектов на вашем компьютере.



Откройте терминал в VS Code (Ctrl+`` или через меню View > Terminal`).

1. Перейдите в папку, где вы хотите создать проект Django. Например:

```bash
cd ~/projects
```

2. Создайте новую папку для вашего проекта:

```bash
mkdir myfirstdjango
cd myfirstdjango
  ```


image.png


3. Создайте виртуальное окружение:

   ```bash
   python -m venv myenv
   ```

   Здесь `myenv` — это имя вашего виртуального окружения.

4. Активируйте виртуальное окружение:

   - На Windows:

     ```bash
     myenv\Scripts\activate
     ```

   - На macOS/Linux:

     ```bash
     source myenv/bin/activate
     ```

   После активации в терминале появится префикс `(myenv)`.


Если при активации виртуального окружения возникает ошибка следующего содержания:

```
myenv\Scripts\activate : File D:\Projects\myfirstdjango\myenv\Scripts\Activate.ps1 cannot be loaded because running scripts is disabled on this system. For more information, see about_Execution_Policies at https:/go.microsoft.com/fwlink/?LinkID=135170.
At line:1 char:1
+ myenv\Scripts\activate
+ ~~~~~~~~~~~~~~~~~~~~~~
    + CategoryInfo          : SecurityError: (:) [], PSSecurityException
    + FullyQualifiedErrorId : UnauthorizedAccess
```

это свидетельствует о том, что выполнение скриптов в системе отключено. В таком случае рекомендуется сначала выполнить команду для проверки текущей политики выполнения:

```powershell
Get-ExecutionPolicy -Scope CurrentUser
```

```powershell
Set-ExecutionPolicy RemoteSigned -Scope CurrentUser
```

После успешного изменения политики выполнения можно повторно попытаться активировать виртуальное окружение с помощью команды:

```powershell
myenv\Scripts\activate
```

Следуя данным рекомендациям, вы сможете устранить возникшую проблему и успешно активировать виртуальное окружение.

### 6. Структура проекта и приложения

После создания проекта и приложения структура вашего проекта будет выглядеть следующим образом:

```
myproject/
│
├── myproject/               # Основная папка проекта (конфигурации и настройки)
│   ├── __init__.py          # Файл, который указывает, что это Python-пакет
│   ├── asgi.py              # Точка входа для ASGI-серверов (асинхронный интерфейс)
│   ├── settings.py          # Основной файл настроек проекта (база данных, статические файлы и т.д.)
│   ├── urls.py              # Основной файл маршрутов (URL-адреса) проекта
│   ├── wsgi.py              # Точка входа для WSGI-серверов (синхронный интерфейс)
│   └── ...                  # Дополнительные файлы (например, локальные настройки)
│
├── myapp/                   # Папка приложения (логика приложения)
│   ├── migrations/          # Папка для хранения миграций базы данных
│   ├── __init__.py          # Файл, который указывает, что это Python-пакет
│   ├── admin.py             # Регистрация моделей для админ-панели Django
│   ├── apps.py              # Конфигурация приложения (например, имя приложения)
│   ├── models.py            # Определение моделей базы данных
│   ├── tests.py             # Тесты для приложения
│   ├── views.py             # Логика обработки запросов (контроллеры)
│   ├── urls.py              # Маршруты (URL-адреса) для этого приложения (опционально)
│   ├── templates/           # Папка для HTML-шаблонов (если используется)
│   └── ...                  # Дополнительные файлы (например, формы, утилиты)
│
├── manage.py                # Утилита для управления проектом (запуск сервера, миграции и т.д.)
└── ...                      # Дополнительные файлы и папки (например, requirements.txt, static, media)
```

#### Подробное описание структуры:

1. **myproject/**  
   Это основная директория проекта, которая содержит глобальные настройки и конфигурации.  
   - **`__init__.py`**: Пустой файл, который указывает Python, что эту директорию следует рассматривать как пакет.  
   - **`asgi.py`**: Точка входа для ASGI-серверов (например, Daphne), используется для асинхронных приложений.  
   - **`settings.py`**: Основной файл настроек проекта. Здесь настраиваются база данных, статические файлы, middleware, установленные приложения и многое другое.  
   - **`urls.py`**: Основной файл маршрутов (URL-адресов) проекта. Здесь определяются глобальные маршруты, которые могут включать маршруты из приложений.  
   - **`wsgi.py`**: Точка входа для WSGI-серверов (например, Gunicorn), используется для синхронных приложений.  

2. **myapp/**  
   Это директория приложения, где содержится основная логика вашего приложения.  
   - **`migrations/`**: Папка для хранения миграций базы данных. Миграции автоматически создаются Django при изменении моделей.  
   - **`__init__.py`**: Указывает, что это Python-пакет.  
   - **`admin.py`**: Здесь регистрируются модели, чтобы они отображались в админ-панели Django.  
   - **`apps.py`**: Содержит конфигурацию приложения, например, имя приложения.  
   - **`models.py`**: Определение моделей базы данных. Модели представляют собой Python-классы, которые Django преобразует в таблицы базы данных.  
   - **`tests.py`**: Файл для написания тестов вашего приложения.  
   - **`views.py`**: Содержит функции или классы, которые обрабатывают запросы и возвращают ответы (например, рендеринг HTML-страниц).  
   - **`urls.py`**: Опциональный файл для определения маршрутов (URL-адресов) конкретного приложения.  
   - **`templates/`**: Папка для хранения HTML-шаблонов, если они используются в приложении.  

3. **manage.py**  
   Это утилита командной строки, которая позволяет управлять проектом. С ее помощью можно:  
   - Запускать сервер разработки (`python manage.py runserver`).  
   - Создавать и применять миграции (`python manage.py makemigrations`, `python manage.py migrate`).  
   - Создавать суперпользователя для доступа к админ-панели (`python manage.py createsuperuser`).  
   - Запускать тесты (`python manage.py test`).  

4. **Дополнительные файлы и папки**  
   - **`requirements.txt`**: Файл для указания зависимостей проекта.  
   - **`static/`**: Папка для хранения статических файлов (CSS, JavaScript, изображения).  
   - **`media/`**: Папка для хранения медиафайлов, загружаемых пользователями (например, изображения, документы).  


### 7. Запуск сервера разработки

1. Убедитесь, что вы находитесь в папке проекта (где находится `manage.py`).
2. Запустите сервер разработки:

   ```bash
   python manage.py runserver
   ```


image.png

Если при запуске сервера разработки Django вы видите сообщение:  

**"You have 18 unapplied migration(s). Your project may not work properly until you apply the migrations for app(s): admin, auth, contenttypes, sessions. Run 'python manage.py migrate' to apply them."**,  

это означает, что в вашем проекте есть **непримененные миграции**. Такая ситуация является нормальной для нового проекта, поскольку Django автоматически создает миграции для встроенных приложений, таких как:  

- **`admin`**: отвечает за административную панель Django.  
- **`auth`**: обеспечивает систему аутентификации и управления пользователями.  
- **`contenttypes`**: отвечает за хранение типов контента в проекте.  
- **`sessions`**: управляет сессиями пользователей.  

Эти приложения необходимы для базовой функциональности Django, и их миграции нужно применить к базе данных, чтобы проект работал корректно.



3. Откройте браузер и перейдите по адресу `http://127.0.0.1:8000/`. Вы увидите страницу приветствия Django.

### Как исправить?
Чтобы применить миграции и устранить это предупреждение, выполните следующую команду в терминале:

```bash
python manage.py migrate
```

### Что делает команда `migrate`?
- Она применяет все непримененные миграции к вашей базе данных.
- Создает необходимые таблицы в базе данных для встроенных приложений Django.
- После выполнения этой команды предупреждение исчезнет, и ваша база данных будет готова к использованию.

### Что произойдет после применения миграций?
- В вашей базе данных появятся таблицы, необходимые для работы встроенных приложений Django.
- Вы сможете использовать административную панель (`/admin`), систему аутентификации и другие функции Django.
- Предупреждение о непримененных миграциях больше не будет появляться при запуске сервера.

### Часть 2: Использование Docker

Теперь, когда у вас есть базовый Django-проект, давайте добавим Docker, чтобы сделать ваше приложение более переносимым и удобным для разработки. Мы создадим структуру проекта, добавим необходимые файлы и настроим Docker для работы с Django.



### Структура проекта

После выполнения первой части у вас должна быть следующая структура проекта:

```
myfirstdjango/
├── myenv/                  # Виртуальное окружение
├── myproject/              # Основной проект Django
│   ├── __init__.py
│   ├── asgi.py
│   ├── settings.py
│   ├── urls.py
│   └── wsgi.py
├── myapp/                  # Приложение Django
│   ├── __init__.py
│   ├── admin.py
│   ├── apps.py
│   ├── migrations/
│   ├── models.py
│   ├── tests.py
│   └── views.py
├── manage.py               # Управляющий скрипт Django
└── requirements.txt        # Файл зависимостей (его мы создадим)
```

Теперь добавим файлы, необходимые для Docker:

```
myfirstdjango/
├── Dockerfile              # Файл для создания Docker-образа
├── docker-compose.yml      # Файл для управления контейнерами
├── myenv/
├── myproject/
├── myapp/
├── manage.py
└── requirements.txt
```



### 1. **Создание `Dockerfile`**

`Dockerfile` — это файл, который содержит инструкции для создания Docker-образа. Создайте его в корневой папке проекта (`myfirstdjango/`):

```Dockerfile
# Используем официальный образ Python
FROM python:3.13-slim

# Устанавливаем рабочую директорию внутри контейнера
WORKDIR /app

# Копируем файл с зависимостями
COPY requirements.txt .

# Устанавливаем зависимости
RUN pip install --no-cache-dir -r requirements.txt

# Копируем весь проект в контейнер
COPY . .

# Команда для запуска сервера Django
CMD ["python", "manage.py", "runserver", "0.0.0.0:8000"]
```



### 2. **Создание `requirements.txt`**

Файл `requirements.txt` содержит список зависимостей вашего проекта. Создайте его в корневой папке проекта (`myfirstdjango/`):

```txt
Django==5.1.5
```

Если вы используете дополнительные библиотеки, добавьте их сюда. Например:

```txt
Django==5.1.5
psycopg2-binary==2.9.5  # Для работы с PostgreSQL
```


Вы также можете автоматически создать файл, выполнив следующую команду:
```txt
pip freeze > requirements.txt
```
### 3. **Создание `docker-compose.yml`**

`docker-compose.yml` позволяет управлять несколькими контейнерами (например, для Django и базы данных). Создайте этот файл в корневой папке проекта (`myfirstdjango/`):

```yaml
version: '3.8'

services:
  web:
    build: .  # Указываем, что нужно использовать Dockerfile в текущей директории
    ports:
      - "8000:8000"  # Пробрасываем порт 8000 на хост
    volumes:
      - .:/app  # Монтируем текущую директорию в контейнер
    command: python manage.py runserver 0.0.0.0:8000  # Команда для запуска сервера
```

### 5. **Добавление базы данных**

Если ваш проект использует базу данных (например, PostgreSQL), вы можете добавить её в `docker-compose.yml`. Обновите файл следующим образом:

```yaml
version: '3.8'

services:
  web:
    build: .
    ports:
      - "8000:8000"
    volumes:
      - .:/app
    depends_on:
      - db  # Указываем, что контейнер зависит от базы данных
    command: python manage.py runserver 0.0.0.0:8000

  db:
    image: postgres:13  # Используем официальный образ PostgreSQL
    environment:
      POSTGRES_USER: myuser
      POSTGRES_PASSWORD: mypassword
      POSTGRES_DB: mydatabase
    ports:
      - "5432:5432"  # Пробрасываем порт базы данных
```

Обновите настройки базы данных в `myproject/settings.py`:

```python
DATABASES = {
    'default': {
        'ENGINE': 'django.db.backends.postgresql',
        'NAME': 'mydatabase',
        'USER': 'myuser',
        'PASSWORD': 'mypassword',
        'HOST': 'db',  # Имя сервиса в docker-compose.yml
        'PORT': '5432',
    }
}
```

Для отладки Django-приложения в контейнере:
- Установите расширение **Python** для VS Code.
- Создайте файл `.vscode/launch.json`:
```json
{
    "version": "0.2.0",
    "configurations": [
        {
            "name": "Django",
            "type": "python",
            "request": "launch",
            "program": "${workspaceFolder}/manage.py",
            "args": ["runserver", "0.0.0.0:8000"],
            "django": true
        }
    ]
}
```

- Используйте многоступенчатую сборку для уменьшения размера образа.

```Dockerfile
COPY requirements.txt .
RUN pip install --no-cache-dir -r requirements.txt
```

- Чтобы остановить контейнеры, выполните:


- Чтобы удалить контейнеры и образы, выполните:

  ```bash
  docker-compose down --rmi all
  ```
