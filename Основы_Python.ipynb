## Основы Python

### Введение в Python

Python — это высокоуровневый язык программирования, который поддерживает разные парадигмы программирования, включая объектно-ориентированное и функциональное. Он известен своей простотой и читабельностью кода, что делает его отличным выбором для начинающих программистов.

### 1. Ввод и вывод

#### 1.1 Ввод

Для получения ввода от пользователя в Python используется функция `input()`. Она принимает строку в качестве аргумента, которая будет отображена в качестве приглашения для пользователя.

**Пример:**
```python
name = input("Введите ваше имя: ")
print(f"Привет, {name}!")
```

#### 1.2 Вывод

Для вывода информации на экран используется функция `print()`. Вы можете выводить строки, числа и другие объекты.

**Пример:**
```python
age = 25
print("Мне", age, "лет.")
```

**Форматированный вывод:**
```python
price = 19.99
print(f"Цена товара: {price:.2f} рублей.")  # Форматирование до двух знаков после запятой
```

### 2. Типы данных

Python поддерживает несколько встроенных типов данных, включая:

- **Целые числа (`int`)**
- **Числа с плавающей точкой (`float`)**
- **Строки (`str`)**
- **Логические значения (`bool`)**

**Пример:**
```python
x = 10         # int
y = 3.14       # float
name = "Python"  # str
is_valid = True  # bool
```

### 3. Математические функции

Python имеет встроенные математические функции, которые можно использовать для выполнения различных математических операций. Для более сложных операций используется модуль `math`.

**Пример:**
```python
import math

a = 16
sqrt_a = math.sqrt(a)  # Квадратный корень
print(f"Квадратный корень из {a}: {sqrt_a}")

pi_value = math.pi
print(f"Значение π: {pi_value}")
```

### 4. Условный оператор `if`

Условный оператор `if` используется для выполнения кода на основе условия.

**Пример:**
```python
age = int(input("Введите ваш возраст: "))

if age < 18:
    print("Вы несовершеннолетний.")
elif age < 65:
    print("Вы совершеннолетний.")
else:
    print("Вы пенсионер.")
```

### 5. Оператор `case`

Python не имеет встроенного оператора `switch`, но можно использовать `if`-`elif` для достижения подобного результата или воспользоваться конструкцией `match` в версиях Python 3.10 и выше.

**Пример с `match`:**
```python
command = input("Введите команду (start/stop): ")

match command:
    case "start":
        print("Система запущена.")
    case "stop":
        print("Система остановлена.")
    case _:
        print("Неизвестная команда.")
```

### 6. Цикл `for`

Цикл `for` используется для итерации по последовательностям (списки, строки и т.д.).

**Пример:**
```python
fruits = ["яблоко", "банан", "вишня"]

for fruit in fruits:
    print(f"Я люблю {fruit}.")
```

### 7. Цикл `while`

Цикл `while` выполняет блок кода, пока условие истинно.

**Пример:**
```python
count = 0

while count < 5:
    print(f"Счетчик: {count}")
    count += 1
```

### 8. Массивы и матрицы

В Python для работы с массивами и матрицами обычно используется библиотека `NumPy`. Массивы могут быть одномерными или многомерными.

**Пример:**
```python
import numpy as np

# Создание одномерного массива
array_1d = np.array([1, 2, 3, 4, 5])
print("Одномерный массив:", array_1d)

# Создание двумерной матрицы
matrix_2d = np.array([[1, 2, 3], [4, 5, 6]])
print("Двумерная матрица:\n", matrix_2d)

# Доступ к элементам
print("Элемент в первой строке и втором столбце:", matrix_2d[0, 1])
```

### 9. Функции и рекурсия

Функции в Python определяются с помощью ключевого слова `def`. Рекурсия — это метод, при котором функция вызывает сама себя.

**Пример функции:**
```python
def greet(name):
    return f"Привет, {name}!"

print(greet("Александр"))
```

**Пример рекурсивной функции (факториал):**
```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

print(f"Факториал 5: {factorial(5)}")
```
